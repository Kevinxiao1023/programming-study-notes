# API   

## 什么是API            

API 是 Application Programming Interface （应用 编程 接口）三个单词的缩写。

通俗易懂的说，API 其实就是别人已经写好的可以实现特定功能的函数。

只需要调用它，传入规定的参数，这个函数就可以实现功能。

我们可以把 API 的诞生用一个小故事来展示出来：

程序员A开发了软件A，程序员B正在研发软件B。有一天，程序员B想要调用软件A的部分功能，但是他又不想从头看一遍软件A的代码，怎么办呢？

程序员A想了一个好主意：我把软件A里你需要的功能打包好，写成函数；你把这个函数放在软件B里，就能直接用我的功能啦～

其中，API 就是程序员A说的那个函数。

其实日常生活中，有很多 API 的场景。

比如：我们在某宝上下单后，店家选用顺风发货， 某宝和顺风作为两家独立的公司，为什么在某宝上查看到物流信息，这就要用到API。

当查看快递信息时，某宝利用顺风提供的 API，可以实时调取信息呈现在自己的网站上。

## 爬虫与API            

很多网站都会提供不同功能的 API，如果网站提供了，我们就不需要去爬取 HTML 代码，可以直接去网站的 API 获取数据。

这样做既减轻了网站的服务器压力，也降低了我们解析网页的难度。

通常，API 一般以 URL 形式存在，我们可以通过访问这些 URL 来获取数据。

# B站弹幕

在B站中，只要视频中有弹幕，就会有一个 cid 参数。cid 用来表示某个视频对应的弹幕池（每个视频都有弹幕上限，弹幕上限取决于视频长度）。

前面我们提到，API 一般以 URL 形式存在的，B站的弹幕 API 的 URL 链接为：
https://comment.bilibili.com/{cid}.xml

看来，只要知道了 cid 参数值，就可以找到视频弹幕数据的位置了。

## 获取cid     

右键“检查”，开发者工具中点击网络，filter框中搜索cid，进入第一个文件，滑到最下面，可以发现cid。     

由于网站工程师会不停的维护网站，如果你根据视频没有找到 cid，请点击【Headers】旁边的【Payload】，就可以找到 cid。

【Payload】可以理解为是服务器返回给请求者的有用的数据。

## 接口链接          

将获取的 cid 参数值放入以下 URL 链接中。

例如：

在浏览器中打开上面的 URL 链接，进入该网页就可以看到视频中所有弹幕数据。

# XML可拓展标记语言

相信你也发现了，刚才打开的 URL 链接中，返回的结果不是一般的网站，而是如图所示的一个文件。这个文件就是 XML 文件。

这个 XML 就是浏览器中存储弹幕的源数据，而我们看到的弹幕是从这个数据中提取出，再展示到视频中的。

## 什么是XML
XML 全称为可扩展标记语言，是一种标记语言。

标记指计算机所能理解的信息符号，有了标记，我们可以通过程序很方便的识别数据的类型和层次结构。

比如图中的 Tom 和 Jerry 是数据内容，`<to>` 和 `<from>` 用来标记 Tom 和 Jerry ，表示他们分别是收件人和发件人。

在之前的课程中，我们学习过HTML。HTML 使用标记来注明文本、图片和其他内容，以便于在 Web 浏览器中显示。

HTML 标记包含固定的标签如 `<head>，<title>，<body>，<p>，<div>，<span>，<img> `等等。

在 XML 中可以定义自己的标签和文档结构，没有固定的标签。

例如右图，代码中的标签没有在 XML 标准中定义过，这些标签是由 XML 文档的创作者（Jerry）编写的。

## XML与HTML的区别
XML 被设计用来传输和存储数据，其焦点是数据的内容。

HTML 被设计用来显示数据，其焦点是数据的外观。

在这里我们对XML只做了简单介绍，想要了解更多的话可以Google一下哦～

# 思路整理

1. 导入requests模块
2. 找到视频弹幕cid，拼接接口链接
3. 向接口链接发送请求
4. 输出获取XML代码

想要获取视频弹幕接口 XML 代码，首先，我们要导入 requests 模块。

将弹幕接口链接以字符串的格式赋值给变量 url。

接着，将变量 url 作为参数，添加进 requests.get() 函数中，再给赋值给变量 response。

然后，使用 .text 属性获取 response 内容，赋值给变量 xml。在使用 print 输出变量 xml，即可获得视频弹幕接口 XML 代码。

由于 API 是网站提供的，可以不用设置 header。

````python
# TODO 使用import导入requests模块
import requests

# TODO 将https://comment.bilibili.com/218710655.xml赋值给变量url
url = "https://comment.bilibili.com/218710655.xml"

# TODO 将变量url作为参数，添加进requests.get()中，给赋值给response
response = requests.get(url)

# TODO 使用.text属性获取response内容，赋值给xml
xml = response.text

# TODO 使用print输出xml
print(xml)
````

## 相关竞品
Yoyo 在网页的右下方发现了相关推荐，这是网页根据视频信息，推送其他博主制作的视频，Yoyo觉得这些博主很可能是潜在的竞品。

Yoyo想要查看这些视频，从而进行分析；她想要获取视频名称和视频链接，这该怎么办呢？通过这段时间的学习，她想到可以使用爬虫来获取相关数据。

解题思路：
1. 导入相关模块，设置headers；
2. 解析网页，使用 find_all() 函数找到推荐视频相关的节点 class = "title";
3. 从每个节点中，获取 title 的属性值和 href 的属性值；
4. 格式化输出视频名称和链接，用冒号隔开，如下所示。        

URL链接：
https://www.bilibili.com/video/BV18V411z7cV?p=1

User-Agent：
"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"

````python
import requests
from bs4 import BeautifulSoup
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"}
url = "https://www.bilibili.com/video/BV18V411z7cV?p=1"
response = requests.get(url, headers = headers)
html = response.text
soup = BeautifulSoup(html, "lxml")
content_all = soup.find_all(class_ = "title")
for content in content_all:
    videoName = content.attrs["title"]
    videoUrl = content.get("href")
    print(f"{videoName}:{videoUrl}")
````

## 组合书籍信息
在之前的百题斩中，我们帮助小倩成功地从豆瓣网页获得了图书名字。这一次，她想要获得更加完整的书籍信息，如下图所示：包括图书名字，评分和简介这三项，并且将他们逐项输出。这应该如何实现呢？

输出格式如下图所示，每项信息之间空格，每本图书之间空行：
```
红楼梦 9.6 都云作者痴，谁解其中味？
活着 9.4 生的苦难与伟大
……
```

URL链接：https://book.douban.com/top250?start=0

User-Agent：Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36

注意：本题需要用到.text知识点，请点击提示进行学习。

.text

如下图所示，我们可以看到包含“红楼梦”书名的节点没有统一的名称，所以我们不能直接使用find_all()函数。因此，我们需要先使用find()函数获取class="pl2"的节点，然后通过.a方法获取其中的a标签，再通过.text方法，获取a标签下的文本内容，就可以获得“红楼梦”的书名啦。


```
# 使用find()获取pl2节点
pl2 = content.find(class_="pl2")
# 使用.text获取pl2中a标签下的的text文本信息
title = pl2.find("a").text
```

````python
# 使用import导入requests模块
import requests

# 从bs4中导入BeautifulSoup模块
from bs4 import BeautifulSoup

# 将URL地址，赋值给变量url
url = "https://book.douban.com/top250?start=0"

# 将User-Agent以字典键对形式赋值给headers
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"}

# 将 url 和 headers参数，添加进requests.get()中，将字典headers传递headers参数，给赋值给response
response = requests.get(url, headers=headers)

# 将服务器响应内容转换为字符串形式，赋值给html
html = response.text

# 使用BeautifulSoup()传入变量html和解析器lxml，赋值给soup
soup = BeautifulSoup(html, "lxml")

# 使用find_all()查询soup中class="item"的节点，赋值给content_all
content_all = soup.find_all(class_="item")

# TODO for循环遍历content_all
for content in content_all:

    # TODO 使用find查询content中class="pl2"的节点(p后面是字母l)，赋值给pl2
    pl2 = content.find(class_ = "pl2")

    # TODO 使用.text获取pl2中a标签下的的text文本信息，赋值给title
    title = pl2.find(name = "a").text

    # TODO 使用replace()去掉title中的换行符"\n"
    title = title.replace("\n", "")

    # TODO 使用replace()去掉title中的空格
    title = title.replace(" ", "")

    # TODO 使用find查询content中class="rating_nums"的节点，赋值给rating_nums
    rating_nums = content.find(class_ = "rating_nums")

    # TODO 获取rating_nums节点中的内容，赋值给rate
    rate = rating_nums.string

    # TODO 使用find查询content中class="inq"的节点，赋值给inq
    inq = content.find(class_ = "inq")

    # TODO 获取inq节点中的内容，赋值给introduction
    introduction = inq.string

    # TODO 使用格式化输出结果，title,rate,introduction两两之间加空格
    print(f"{title} {rate} {introduction}")
````
