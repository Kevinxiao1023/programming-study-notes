## è·å–å•ä¸ªèŠ‚ç‚¹

ä¹‹å‰æˆ‘ä»¬å­¦ä¹ è¿‡find_all()å‡½æ•°ï¼Œè¯¥å‡½æ•°èƒ½å¤ŸåŒ¹é…æ‰€æœ‰æ»¡è¶³æ¡ä»¶çš„å…ƒç´ ï¼Œå¹¶ä¸”è¿”å›ä¸€ä¸ªåˆ—è¡¨ã€‚

**find()å‡½æ•°åªè¿”å›ç¬¬ä¸€ä¸ªåŒ¹é…åˆ°çš„å…ƒç´ ï¼Œfind()å‡½æ•°çš„ç”¨æ³•ä¸find_all()å‡½æ•°åŸºæœ¬ç±»ä¼¼**ã€‚

æŸ¥æ‰¾class="pic" æ‰€åœ¨çš„æ‰€æœ‰èŠ‚ç‚¹ï¼Œå°±ä¼ å…¥å‚æ•°class_="pic"ï¼ŒæŸ¥æ‰¾<`li></li`>æ ‡ç­¾ï¼Œå°±ä¼ å…¥å‚æ•°name="li"ã€‚

æ¯ä¸ªå›¾ç‰‡å¯¹åº”çš„divå…ƒç´ å†…ï¼Œåªæœ‰ä¸€ä¸ªimgæ ‡ç­¾ï¼Œè¦è·å–è¯¥æ ‡ç­¾ï¼Œä½¿ç”¨find()å‡½æ•°ï¼Œå°†name="img"ä¼ å…¥å‡½æ•°ä¸­ï¼Œå°±èƒ½å¤Ÿæ‹¿åˆ°imgå¯¹åº”çš„æ ‡ç­¾å†…å®¹ã€‚

æ¥ä¸‹æ¥ï¼Œä½¿ç”¨find()å‡½æ•°è·å–imgæ ‡ç­¾å†…å®¹ï¼Œå¹¶è¾“å‡ºã€‚

````python
# ä½¿ç”¨importå¯¼å…¥requestsæ¨¡å—
import requests

# ä½¿ç”¨from..importä»bs4æ¨¡å—å¯¼å…¥BeautifulSoup
from bs4 import BeautifulSoup

# å°†ç”µå½±URLåœ°å€ï¼Œèµ‹å€¼ç»™å˜é‡url
url = "https://movie.douban.com/top250"

# å°†User-Agentä»¥å­—å…¸é”®å¯¹å½¢å¼èµ‹å€¼ç»™headers
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"}

# å°†å­—å…¸headersä¼ é€’ç»™headerså‚æ•°ï¼Œæ·»åŠ è¿›requests.get()ä¸­ï¼Œèµ‹å€¼ç»™response
response = requests.get(url, headers=headers)

# å°†æœåŠ¡å™¨å“åº”å†…å®¹è½¬æ¢ä¸ºå­—ç¬¦ä¸²å½¢å¼ï¼Œèµ‹å€¼ç»™html
html = response.text

# ä½¿ç”¨BeautifulSoup()ä¼ å…¥å˜é‡htmlå’Œè§£æå™¨lxmlï¼Œèµ‹å€¼ç»™soup
soup = BeautifulSoup(html, "lxml")

# ä½¿ç”¨find_all()æŸ¥è¯¢soupä¸­class="pic"çš„èŠ‚ç‚¹ï¼Œèµ‹å€¼ç»™content_all
content_all = soup.find_all(class_="pic")

# forå¾ªç¯éå†content_all
for content in content_all:

    # TODO ä½¿ç”¨find()æŸ¥è¯¢contentä¸­çš„imgæ ‡ç­¾ï¼Œå¹¶èµ‹å€¼ç»™imgContent
    imgContent = content.find(name="img")

    # TODO ä½¿ç”¨printè¾“å‡ºimgContent
    print(imgContent)
````

## è·å–å±æ€§å€¼

å¯¹`<img>`æ ‡ç­¾ä¸­çš„srcå’Œaltå±æ€§ä½¿ç”¨.attrsï¼Œå°±å¯ä»¥è·å–å±æ€§å€¼ã€‚

ä¾‹å¦‚è·å–å›¾ç‰‡æ ‡ç­¾ä¸­çš„å›¾ç‰‡æè¿°å°±å¯ä»¥ä½¿ç”¨ï¼Œ`attrs["alt"]`ï¼Œè·å–å›¾ç‰‡æ ‡ç­¾ä¸­çš„é“¾æ¥å°±å¯ä»¥ä½¿ç”¨`attrs["src"]`ã€‚     

äº†è§£.attrså±æ€§çš„ä½¿ç”¨æ–¹æ³•åï¼Œä½¿ç”¨è¯¥å±æ€§è·å–å›¾ç‰‡æ ‡ç­¾ä¸­çš„é“¾æ¥ï¼Œå¹¶è¾“å‡ºæŸ¥çœ‹ã€‚

````python
# TODO ä½¿ç”¨importå¯¼å…¥requestsæ¨¡å—
import requests

# TODO ä½¿ç”¨from...importä»bs4æ¨¡å—ä¸­å¯¼å…¥BeautifulSoup
from bs4 import BeautifulSoup

# å°†User-Agentä»¥å­—å…¸é”®å¯¹å½¢å¼èµ‹å€¼ç»™headers
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"}

# TODO å°†URL"https://www.futta.net/photo-season/autumn/"èµ‹å€¼ç»™å˜é‡url
url = "https://www.futta.net/photo-season/autumn/"

# TODO ä½¿ç”¨get()å‡½æ•°è¯·æ±‚é“¾æ¥ï¼Œå¹¶ä¸”å¸¦ä¸Šheaders
response = requests.get(url, headers = headers)

# TODO ä½¿ç”¨.textå±æ€§å°†æœåŠ¡å™¨ç›¸åº”å†…å®¹è½¬æ¢ä¸ºå­—ç¬¦ä¸²å½¢å¼ï¼Œèµ‹å€¼ç»™html
html = response.text

# TODO ä½¿ç”¨BeautifulSoup()ä¼ å…¥å˜é‡htmlå’Œè§£æå™¨lxmlï¼Œèµ‹å€¼ç»™soup
soup = BeautifulSoup(html, "lxml")

# TODO ä½¿ç”¨find()æŸ¥è¯¢class="contentsset"å±æ€§ï¼Œå¹¶èµ‹å€¼ç»™content_all
content_all = soup.find(class_="contentsset")

# TODO ä½¿ç”¨find_all()æŸ¥è¯¢liæ ‡ç­¾ï¼Œå¹¶èµ‹å€¼ç»™lis
lis = content_all.find_all(name = "li")
# è®¾ç½®ä¸€ä¸ªè®¡æ•°å™¨countåˆå§‹å€¼ä¸º1
count = 1

# TODO forå¾ªç¯éå†lis
for li in lis:

    # TODO ä½¿ç”¨find()æŸ¥è¯¢imgæ ‡ç­¾ï¼Œå¹¶èµ‹å€¼ç»™imgContent
    imgContent = li.find(name = "img")

    # TODO ä½¿ç”¨ifåˆ¤æ–­ï¼Œå½“è·å–åˆ°çš„imgæ ‡ç­¾å†…å®¹ä¸ºNoneæ—¶ï¼Œå°±è·³è¿‡
    if imgContent is None:
        continue
    
    # TODO ä½¿ç”¨.attrså±æ€§è·å–srcå¯¹åº”çš„å±æ€§å€¼ï¼Œå¹¶èµ‹å€¼ç»™imgUrl
    imgUrl = imgContent.attrs["src"]
    
    # TODO ä½¿ç”¨get()å‡½æ•°è¯·æ±‚é“¾æ¥ï¼Œå¹¶èµ‹å€¼ç»™imgResponse
    imgResponse = requests.get(imgUrl)

    # TODO ä½¿ç”¨.contentå±æ€§å°†å“åº”æ¶ˆæ¯è½¬æ¢æˆå›¾ç‰‡æ•°æ®ï¼Œèµ‹å€¼ç»™img
    img = imgResponse.content

    # TODO ä½¿ç”¨with open()ä»¥å›¾ç‰‡å†™å…¥çš„æ–¹å¼æ‰“å¼€æ–‡ä»¶
    # ç”¨æ ¼å¼åŒ–å°†countå’Œ.jpgæ ¼å¼ç»„åˆ
    # æ‰“å¼€çš„æ–‡ä»¶èµ‹å€¼ä¸ºf
    with open(f"/Users/xiaojiangyue/programming-study-notes/Python/çˆ¬è™«/lesson_9_output/ç§‹å¶/{count}.jpg", "wb") as f:
        # TODO ä½¿ç”¨write()å‡½æ•°å†™å…¥å›¾ç‰‡
        f.write(img)
    
    # TODO å®Œæˆä¸€å¼ å›¾ç‰‡å†™å…¥æ—¶ï¼Œcountè®¡æ•°åŠ 1
    count += 1
````

# ä½¿ç”¨Imageæ¨¡å—ä¸­çš„filterå‡½æ•°ï¼Œä¼ å…¥å‚æ•°Imagefilter.EMBOSSä¸ºå›¾ç‰‡æ·»åŠ æ»¤é•œ

````python
# TODO ä½¿ç”¨importå¯¼å…¥requestsæ¨¡å—
import requests
# TODO ä½¿ç”¨from..importä»bs4æ¨¡å—å¯¼å…¥BeautifulSoup
from bs4 import BeautifulSoup
# TODO ä½¿ç”¨from..importä»PILæ¨¡å—å¯¼å…¥Image,ImageFilter
from PIL import Image, ImageFilter

# å°†ç”µå½±URLåœ°å€ï¼Œèµ‹å€¼ç»™å˜é‡url
url = "https://nocturne-spider.baicizhan.com/2020/08/20/photo/"
# å°†User-Agentä»¥å­—å…¸é”®å¯¹å½¢å¼èµ‹å€¼ç»™headers
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"}
# å°†å­—å…¸headersä¼ é€’ç»™headerså‚æ•°ï¼Œæ·»åŠ è¿›requests.get()ä¸­ï¼Œèµ‹å€¼ç»™response
response = requests.get(url, headers=headers, verify=False)
# å°†æœåŠ¡å™¨å“åº”å†…å®¹è½¬æ¢ä¸ºå­—ç¬¦ä¸²å½¢å¼ï¼Œèµ‹å€¼ç»™html
html = response.text
# ä½¿ç”¨BeautifulSoup()ä¼ å…¥å˜é‡htmlå’Œè§£æå™¨lxmlï¼Œèµ‹å€¼ç»™soup
soup = BeautifulSoup(html, "lxml")
# TODO ä½¿ç”¨find()æŸ¥è¯¢soupä¸­class="post-body"çš„èŠ‚ç‚¹ï¼Œèµ‹å€¼ç»™content_all
content_all = soup.find(class_="post-body")
# TODO ä½¿ç”¨find_all()æŸ¥è¯¢content_allä¸­çš„"img"èŠ‚ç‚¹ï¼Œèµ‹å€¼ç»™img_all
img_all = content_all.find_all(name = "img")

# åˆ›å»ºè®°å½•å›¾ç‰‡åå­—çš„å˜é‡imgNameå¹¶èµ‹å€¼ä¸º1
imgName = 1

# TODO forå¾ªç¯éå†img_all
for img_content in img_all:
    
    # TODO ä½¿ç”¨.attrsè·å–srcå¯¹åº”çš„å±æ€§å€¼ï¼Œå¹¶èµ‹å€¼ç»™imgUrl
    imgUrl = img_content.attrs["src"]
    # TODO ä½¿ç”¨requests.get()è¯·æ±‚å›¾ç‰‡é“¾æ¥ï¼Œèµ‹å€¼ç»™imgResponse
    imgResponse = requests.get(imgUrl, verify=False, headers=headers)
    # TODO ä½¿ç”¨.contentå±æ€§å°†å“åº”æ¶ˆæ¯è½¬æ¢æˆå›¾ç‰‡æ•°æ®ï¼Œèµ‹å€¼ç»™img
    img = imgResponse.content

    # TODO ä½¿ç”¨with...asè¯­å¥é…åˆopen()å‡½æ•°ä»¥"wb"æ–¹å¼æ‰“å¼€æ–‡ä»¶
    # ç”¨æ ¼å¼åŒ–å°†å›¾ç‰‡åå­—å’Œ.jpgæ ¼å¼ç»„åˆ
    with open(f"{imgName}.jpg", "wb") as f:
        # TODO ä½¿ç”¨write()å°†å›¾ç‰‡å†™å…¥
        f.write(img)

    # TODO ä½¿ç”¨Imageæ¨¡å—ä¸­çš„openå‡½æ•°æ‰“å¼€å›¾ç‰‡æ–‡ä»¶ï¼Œèµ‹å€¼ç»™img_before
    # ä½¿ç”¨æ ¼å¼åŒ–å°†å›¾ç‰‡åå­—å’Œ.jpgæ ¼å¼ç»„åˆ
    img_before = Image.open(f"{imgName}.jpg")
    # TODO ä½¿ç”¨filter()å‡½ï¼Œè°ƒç”¨å‚æ•°ImageFilter.EMBOSSï¼Œèµ‹å€¼ç»™img_after
    img_after = img_before.filter(ImageFilter.EMBOSS)
    # TODO ä½¿ç”¨save()å‡½æ•°ä¿å­˜æ”¹åçš„å›¾ç‰‡
    img_after.save(f"{imgName}.jpg")
    # å°†è®°å½•å›¾ç‰‡åå­—çš„å˜é‡imgNameå€¼å¢åŠ 1
    imgName = imgName+1
````
