# 函数的定义与调用

在上节课中我们从网页中，获取到了 100 条职位链接，每条链接中的网页都包含了职位名称、公司名称、所在城市与薪资。

在这里，我们需要针对每一个职位详情链接都要做类似的操作（获取网页信息、解析并提取网页中的职位名称、公司名称、所在城市与薪资）。

这种需要重复执行多次的相似代码块就很适合写成函数，因为一个好的程序员需要遵循一条原则：“Don’t Repeat Yourself”，简称 DRY 原则。

在这里，我们可以定义一个函数，命名为 getPositionInfo，参数为请求的 URL。

这个函数的功能是请求指定的 URL 并获取网页内容，使用 BeautifulSoup 解析网页，最后提取出需要的数据。

好了，让我们来编写函数定义与调用的代码吧。首先，定义一个新函数 getPositionInfo，传入参数 detail_url；

将 detail_url 和 headers 作为参数，添加进 requests.get() 函数中，赋值给变量 res；

接着，使用 .text 属性获取网页内容，赋值给 html，由于网页内容较多，我们这里可以用切片的方法，先输出前100个字符；

用 BeautifulSoup() 函数创建一个 BeautifulSoup 对象传入变量 html 和解析器 lxml，赋值给 soup。

然后在获取职位对应的 URL 链接后，使用 getPositionInfo 函数，将该获取到的每个链接作为参数传入。

````python
# 导入requests模块
import requests

# 从bs4中导入BeautifulSoup模块
from bs4 import BeautifulSoup

# 导入time模块
import time

# 将User-Agent以字典键对形式赋值给headers
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"}

# TODO 定义一个新函数getPositionInfo，包含参数detail_url
def getPositionInfo(detail_url):
    
    # TODO 将detail_url和headers参数，添加进requests.get()中，给赋值给res
    res = requests.get(detail_url, headers = headers)

    # TODO 使用.text属性获取网页前100个字符的内容，赋值给html
    html = res.text[:100]

    # TODO 用BeautifulSoup()传入变量html和解析器lxml，赋值给soup
    soup = BeautifulSoup(html, "lxml")    

    # TODO print输出soup
    print(soup)

# for循环遍历range()函数生成的1-6的数字
for i in range(1,6):

    # 利用格式化字符生成串网站链接 赋值给变量url
    url = f"https://www.shixiseng.com/interns?page={i}&type=intern&keyword=%E4%BA%A7%E5%93%81%E7%BB%8F%E7%90%86&area=&months=&days=&degree=&official=entry&enterprise=&salary=-0&publishTime=&sortType=&city=%E5%85%A8%E5%9B%BD&internExtend="
    
    # 将url和headers参数，添加进requests.get()中，将字典headers传递给headers参数，给赋值给res
    res = requests.get(url, headers=headers)
    
    # 使用.text属性获取网页内容，赋值给html
    html = res.text

    # 用BeautifulSoup()传入变量html和解析器lxml，赋值给soup
    soup = BeautifulSoup(html,"lxml")
    
    # 使用find_all()查询soup中class=title ellipsis font的节点，赋值给titles
    titles = soup.find_all(class_ = "title ellipsis font")

    # for循环遍历列表titles
    for item in titles:

        # 使用.attrs获取href对应的属性值，并赋值给detail_url
        detail_url = item.attrs["href"]

        # TODO 调用getPositionInfo()函数，传入参数detail_url
        getPositionInfo(detail_url)
    
    # 使用time.sleep()停顿2秒
    time.sleep(2)
````

在输出的结果中，日薪数字还是以方格形式呈现出来，看来我们遇到了字体反爬虫。

所谓字体反爬虫，就是一些关键数据在网页上查看时，它是正常显示的；而当查看网页的 HTML 代码时，却呈现出一个个的方块。

这样我们就没办法将网页上的数据直接爬取下来。

虽然有字体反爬虫，我们也有办法破解它。HTML 代码中的小方格与网页中数字的对应一定存在某种规则。

这种规则是网站自己的一套编码方式，在这里我们可以绕道而行。

## 替换字体
在这里，我们可以通过一种特殊的字体，将小方格显示成字符。

找到小方格原数字与字符之间的关系后，统一替换成正常的数字。

这样不就避开了网页中的字体反爬虫了么

我们可以先把小方格用其他的编码方式重新编码，比如说“UTF-8”编码，。

找到编码后的二进制数据与网页薪资数字之间的对应关系，然后将它们进行替换。

接着将二进制数据，再转换成字符串，就可以破解字体反爬虫机制，获取到薪资数据啦～

# 字符串编码转换

## encode编码
想要将字符串转换成二进制数据，在这里我们可以使用 encode() 函数，实现“编码”。

在 encode() 函数中默认 UTF-8 编码，由于它是针对英语设计的，中英文字符转码后会有差异。

以代码为例，字母和数字经过编码后格式不变；每个中文字符对应三个“\”分割的字符内容。

````python
alp = "yequ"
num = "666"
word = "夜"
words = "夜曲" 
print(alp.encode())
print(num.encode())
print(word.encode())
print(words.encode())
````

接下来，我们对某个职位薪资进行编码，需要重新编写代码获取网页源代码，解析网页提取出薪资。

大部分的代码相信你已经很熟悉了，接着需要使用 encode() 函数对网页提取出薪资小方块进行编码，再使用 print 输出。

````python
# 导入requests模块
import requests

# 从bs4中导入BeautifulSoup模块
from bs4 import BeautifulSoup

# 将User-Agent以字典键对形式赋值给headers
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"}

# 将URL链接以字符串格式赋值给url
url = "https://www.shixiseng.com/intern/inn_bzdfqbshaem3?pcm=pc_SearchList"

# 将url和headers参数，添加进requests.get()中，将字典headers传递给headers参数，给赋值给res
res = requests.get(url,headers=headers)

# 使用.text属性获取网页内容，赋值给html
html = res.text

# TODO 用BeautifulSoup()传入变量html和解析器lxml，赋值给soup
soup = BeautifulSoup(html, "lxml")

# TODO 使用find()函数获取class="job_money cutom_font"的节点
# 使用.string属性提取出标签内容，赋值给salary
salary = soup.find(class_="job_money cutom_font").string

# TODO encode()函数对变量salary编码，赋值给salary
salary = salary.encode()

# TODO 使用print输出salary
print(salary)
````

## 对应关系
“120-200/天”经过编码后转换成二进制数据，他们的对应关系是什么样的呢？

将网页上显示的内容和最后拿到的二进制数据进行对比，以短横线 “-” 和反斜杠 “/” 进行拆分的话，对应关系应该如图所示。

注意，网站的反爬虫机制会让这种对应关系不定时的更新。

好了，我们找到了数字 0、1、2 对应的二进制数据，接下来，我们要用对二进制数据进行替换，把它们替换成相应的 UTF-8 编码数字。

* b"0" 替换 b"\xee\x8b\xbf"
* b"1" 替换 b"\xee\xa2\x9c"

````python
# 导入requests模块
import requests

# 从bs4中导入BeautifulSoup模块
from bs4 import BeautifulSoup

# 导入time模块
import time

# 将User-Agent以字典键对形式赋值给headers
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"}

# 定义一个新函数getPositionInfo，包含参数detail_url
def getPositionInfo(detail_url):
    
    # 将detail_url和headers参数，添加进requests.get()中，给赋值给res
    res = requests.get(detail_url,headers=headers)

    # 使用.text属性获取网页内容，赋值给html
    html = res.text

    # 用BeautifulSoup()传入变量html和解析器lxml，赋值给soup
    soup = BeautifulSoup(html, "lxml")

    # TODO 使用find()函数获取class="new_job_name"的节点
    # 使用attrs属性提取出title的属性值,赋值给变量job
    job = soup.find(class_ = "new_job_name")
    job_title = job.find(name = "span").string.strip()

    # TODO 使用find()函数获取class="com-name"的节点
    # 使用.string属性提取出标签内容，赋值给companyName
    # 使用strip()移除换行符，赋值给companyName
    companyName = soup.find(class_ = "com-name").string.strip()

    # TODO 使用find()函数获取class="job_position"的节点
    # 使用.string属性提取出标签内容，赋值给position
    position = soup.find(class_ = "job_position").string
    
    # TODO 使用find()函数获取class="job_money cutom_font"的节点
    # 使用.string属性提取出标签内容，赋值给salary
    salary = soup.find(class_="job_money cutom_font").string

    
    # TODO 使用encode()函数对变量salary编码，赋值给salary
    salary = salary.encode()

    # 使用replace()函数将二进制数据替换成UTF-8编码数字0-2
    salary = salary.replace(b"\xee\x8b\xbf", b"0")
    salary = salary.replace(b"\xee\xa2\x9c", b"1")
    salary = salary.replace(b"\xee\x90\xb7", b"2")
    # 使用replace()函数将二进制数据替换成UTF-8编码数字3-6
    salary = salary.replace(b"\xee\x81\xa5", b"3")
    salary = salary.replace(b"\xee\xad\xb1", b"4")
    salary = salary.replace(b"\xee\xb2\xae", b"5")
    salary = salary.replace(b"\xef\x8a\x98", b"6")
    # TODO 使用replace()函数将二进制数据替换成UTF-8编码数字7-9
    salary = salary.replace(b"\xef\x80\xa6", b"7")
    salary = salary.replace(b"\xee\xa1\xb1", b"8")
    salary = salary.replace(b"\xee\xbe\xad", b"9")
    # TODO 使用decode()函数对替换后的二进制数据进行解码，赋值给salary
    salary = salary.decode()

    # TODO 使用print格式化输出job,companyName,position,salary
    print(f"{job_title},{companyName},{position},{salary}")

# for循环遍历range()函数生成的1-6的数字
for i in range(1,6):

    # 利用格式化字符生成串网站链接 赋值给变量url
    url = f"https://www.shixiseng.com/interns?page={i}&type=intern&keyword=%E4%BA%A7%E5%93%81%E7%BB%8F%E7%90%86&area=&months=&days=&degree=&official=entry&enterprise=&salary=-0&publishTime=&sortType=&city=%E5%85%A8%E5%9B%BD&internExtend="
    
    # 将url和headers参数，添加进requests.get()中，将字典headers传递给headers参数，给赋值给res
    res = requests.get(url, headers=headers)
    
    # 使用.text属性获取网页内容，赋值给html
    html = res.text

    # 用BeautifulSoup()传入变量html和解析器lxml，赋值给soup
    soup = BeautifulSoup(html,"lxml")
    
    # 使用find_all()查询soup中class=title ellipsis font的节点，赋值给titles
    titles = soup.find_all(class_ = "title ellipsis font")

    # for循环遍历列表titles
    for item in titles:

        # 使用.attrs获取href对应的属性值，并赋值给detail_url
        detail_url = item.attrs["href"]

        # 调用getPositionInfo()函数，传入参数detail_url
        getPositionInfo(detail_url)
    
    # 使用time.sleep()停顿2秒
    time.sleep(2)
````

````python
# 导入requests模块
import requests

# 从bs4中导入BeautifulSoup模块
from bs4 import BeautifulSoup

# 导入time模块
import time

# 将User-Agent以字典键对形式赋值给headers
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.102 Safari/537.36"}


# 定义一个新函数getPositionInfo，包含参数detail_url
def getPositionInfo(detail_url):
    # 将detail_url和headers参数，添加进requests.get()中，给赋值给res
    res = requests.get(detail_url, headers=headers)

    # 使用.text属性获取网页内容，赋值给html
    html = res.text

    # 用BeautifulSoup()传入变量html和解析器lxml，赋值给soup
    soup = BeautifulSoup(html, "lxml")

    # TODO 使用find()函数获取class="new_job_name"的节点
    # 使用attrs属性提取出title的属性值,赋值给变量job
    job = soup.find(class_ = "new_job_name")
    job_title = job.find(name = "span").string.strip()
    
    # TODO 使用find()函数获取class="com-name"的节点
    # 使用.string属性提取出标签内容
    # 使用strip()移除空格，赋值给companyName
    companyName = soup.find(class_="com-name").string.strip()

    # TODO 使用find()函数获取class="job_position"的节点
    # 使用.string属性提取出标签内容，赋值给position
    position = soup.find(class_="job_position").string


    # TODO 使用find()函数获取class="job_time cutom_font"的节点
    # 使用.string属性提取出标签内容，赋值给month
    month = soup.find(class_="job_time cutom_font").string

    # TODO 使用encode()函数对变量month编码，赋值给month
    month = month.encode()

    # 多次使用replace()函数将二进制数据替换成UTF-8编码数字0-5
    month = month.replace(b"\xee\x8b\xbf", b"0").replace(b"\xee\xa2\x9c", b"1").replace(b"\xee\x90\xb7",b"2")
    month = month.replace(b"\xee\x81\xa5",b"3").replace(b"\xee\xad\xb1", b"4").replace(b"\xee\xb2\xae", b"5")

    # TODO 多次使用replace()函数将二进制数据替换成UTF-8编码数字6-9
    month = month.replace(b"\xef\x8a\x98", b"6").replace(b"\xef\x80\xa6", b"7").replace(b"\xee\xa1\xb1", b"8").replace(b"\xee\xbe\xad", b"9")

    # TODO 使用decode()函数对替换后的二进制数据进行解码，赋值给month
    month = month.decode()

    # TODO 格式化输出job,companyName,position,month，每个变量之间用","隔开
    print(f"{job_title},{companyName},{position},{month}")
# for循环遍历range()函数生成的1-3的数字
for i in range(1, 4):

    # 利用格式化字符生成串网站链接 赋值给变量url
    url = f"https://www.shixiseng.com/interns?page={i}&type=intern&keyword=%E4%BA%A7%E5%93%81%E7%BB%8F%E7%90%86&area=&months=&days=&degree=&official=entry&enterprise=&salary=-0&publishTime=&sortType=&city=%E5%85%A8%E5%9B%BD&internExtend="

    # 将url和headers参数，添加进requests.get()中，将字典headers传递给headers参数，给赋值给res
    res = requests.get(url, headers=headers)

    # 使用.text属性获取网页内容，赋值给html
    html = res.text

    # 用BeautifulSoup()传入变量html和解析器lxml，赋值给soup
    soup = BeautifulSoup(html, "lxml")

    # 使用find_all()查询soup中class=title ellipsis font的节点，赋值给titles
    titles = soup.find_all(class_="title ellipsis font")

    # for循环遍历列表
    for item in titles:
        # 使用.attrs获取href对应的属性值，并赋值给detail_url
        detail_url = item.attrs["href"]

        # 调用getPositionInfo()函数，传入参数detail_url
        getPositionInfo(detail_url)

    # 使用time.sleep()停顿2秒
    time.sleep(2)
````
